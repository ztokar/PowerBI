```python
import pandas as pd
pd.read_csv("/content/drive/MyDrive/value_inc_seasons.csv")
data=pd.read_csv("/content/drive/MyDrive/transaction.csv",sep=";")
data.head()
```

```python
#calculating cost, sales price, profit, and mark up per transactiom
#adding columns
data["Cost Per Transaction"]=data['CostPerItem']*data['NumberOfItemsPurchased']
data["Selling Price per Transaction"]=data['SellingPricePerItem']*data['NumberOfItemsPurchased']
data["Profit per Transaction"]=data['Selling Price per Transaction']-data['Cost Per Transaction']
data["Mark up"]=data['Profit per Transaction']/data['Cost Per Transaction']
data["Mark up"]=round(data['Mark up'],2)
data.head()

#fixing up time columns into one column and converting to time data type
#converting to string to concatanate and then to_datetime
year=data["Year"].astype(str)
month=data["Month"].astype(str)
day=data["Day"].astype(str)
fullday=year+"-"+month+"-"+day
fullday.head()
data["Revised Date"]=pd.to_datetime(fullday)
data.head()
```

```python
#splitting the client keyword  coliumn
data["ClientKeywords"]=data['ClientKeywords'].str.replace("[",'')
data["ClientKeywords"]=data['ClientKeywords'].str.replace("]",'')

Client_split=data["ClientKeywords"].str.split(",",expand=True)
Client_split.head()
data["CLient age"]=Client_split[0]
data["CLient type"]=Client_split[1]
data["CLient contract"]=Client_split[2]
data.head()
```

```python
#editing description column
data["ItemDescription"]=data["ItemDescription"].str.lower()
data['ItemDescription']
```

```python
# bringing in season file in preparation to merge
val_seasons=pd.read_csv("/content/drive/MyDrive/value_inc_seasons.csv",sep=';')
val_seasons.head()
```

```python
data= pd.merge(data, val_seasons, on='Month')
data.info()
```
